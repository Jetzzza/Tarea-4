#**Ejercicio 1**:
Diseña un programa que convierta dólares americanos a pesos mexicanos, euros, libras
esterlinas, yuanes y rublos utilizando una tasa de cambio fija. Muestre la tasa de cambio e imprima los cinco resultados.


```python
# Tasas de cambio fijas
tasa_mxn = 18.50  # Pesos Mexicanos por Dólar
tasa_eur = 0.85   # Euros por Dólar
tasa_lib = 0.75   # Libras Esterlinas por Dólar
tasa_yun = 6.50   # Yuanes por Dólar
tasa_rub = 75.00  # Rublos por Dólar

# Función de conversión
def convertir_dolares(cantidad_usd):
    mxn = cantidad_usd * tasa_mxn
    eur = cantidad_usd * tasa_eur
    lib = cantidad_usd * tasa_lib
    yun = cantidad_usd * tasa_yun
    rub = cantidad_usd * tasa_rub
    return mxn, eur, lib, yun, rub

# Solicitar cantidad al usuario
cantidad_usd = float(input("Ingrese una cantidad en dólares: "))

# Realizar la conversión
mxn, eur, lib, yun, rub = convertir_dolares(cantidad_usd)

# Mostrar resultados
print(f"Tasas de cambio (fijas): \n1 USD = {tasa_mxn} MXN\n1 USD = {tasa_eur} EUR\n1 USD = {tasa_lib} Libras Esterlinas\n1 USD = {tasa_yun} YUANES\n1 USD = {tasa_rub} RUBLOS")
print(f"\n {cantidad_usd} USD Valen:")
print(f"{mxn:.2f} Pesos Mexicanos")
print(f"{eur:.2f} Euros")
print(f"{lib:.2f} Libras Esterlinas")
print(f"{yun:.2f} Yuanes")
print(f"{rub:.2f} Rublos")

```

#**Ejercicio 3**:
Desarrolla un programa que calcule el precio final de un producto después de aplicar un descuento.
 El usuario debe ingresar el precio original y el porcentaje de descuento. Imprima el precio final.

```python
# Función para calcular el precio final con descuento
def calcular_precio_final(precio_original, porcentaje_descuento):
    descuento = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - descuento
    return precio_final

# Solicitar el precio original y el porcentaje de descuento al usuario
precio_original = float(input("Ingrese el precio original del producto: "))
porcentaje_descuento = float(input("Ingrese el porcentaje de descuento: "))

# Calcular el precio final
precio_final = calcular_precio_final(precio_original, porcentaje_descuento)

# Mostrar el resultado
print(f"El precio final después de aplicar un {porcentaje_descuento}% de descuento es: ${precio_final:.2f}")
```

#**Ejercicio 5**:
Programe una función que resuelva ecuaciones de segundo grado utilizando la fórmula general.
El usuario debe ingresar los valores de a, b y c. Genere valores por default por si el usuario no ingresa
un número entero o flotante. Además, calcule el valor del discriminante
(b^2 - 4ac) y determine si tiene soluciones reales, soluciones complejas. Imprima el resultado.

```python
import cmath

# Función para resolver ecuaciones de segundo grado
def resolver_ecuacion_segundo_grado(a=1, b=1, c=0):
    discriminante = b**2 - 4*a*c
    if discriminante >= 0:
        raiz1 = (-b + discriminante**0.5) / (2*a)
        raiz2 = (-b - discriminante**0.5) / (2*a)
        tipo_soluciones = "reales"
    else:
        raiz1 = (-b + cmath.sqrt(discriminante)) / (2*a)
        raiz2 = (-b - cmath.sqrt(discriminante)) / (2*a)
        tipo_soluciones = "complejas"

    return discriminante, raiz1, raiz2, tipo_soluciones

# Solicitar valores de a, b y c al usuario
try:
    a = float(input("Ingrese el valor de 'a' (default es 1): ") or 1)
    b = float(input("Ingrese el valor de 'b' (default es 1): ") or 1)
    c = float(input("Ingrese el valor de 'c' (default es 0): ") or 0)
except ValueError:
    print("Se ingresó un valor no válido. Usando valores por defecto: a=1, b=1, c=0")
    a, b, c = 1, 1, 0

# Calcular soluciones
discriminante, raiz1, raiz2, tipo_soluciones = resolver_ecuacion_segundo_grado(a, b, c)

# Mostrar resultados
print(f"El valor del discriminante es: {discriminante}")
print(f"Las soluciones de la ecuación son {tipo_soluciones}: {raiz1} y {raiz2}")
```

#**Ejercicio 6**

 Escriba una función que reciba una lista de números y devuelva dos listas: una con los números positivos y otra con los negativos.

```python
# Función para separar números positivos y negativos
def separar_numeros(lista):
    positivos = []
    negativos = []

    for num in lista:
        if num >= 0:
            positivos.append(num)
        else:
            negativos.append(num)

    return positivos, negativos

# Ejemplo de uso
numeros = [3, -1, 4, -2, 0, -5, 6]
positivos, negativos = separar_numeros(numeros)

print("Números positivos:", positivos)
print("Números negativos:", negativos)
```

#**Ejercicio 7**

Escriba una función que reciba una lista de datos (la de abajo) y devuelva listas con los datos filtrados. Este programa debe separar los datos en **int**, **str**, **bool**, **float**. **Hint**: use la función **append** para insertar datos en las listas.

```python
Datos= [5, 2.4, "hola", 7, 10, 3.14,True,False, "Hi", 0, 0.265]# lista a usar
```

```python
# Función para separar datos por tipo, mejorada para distinguir booleanos de cadenas
def separar_por_tipo(lista):
    enteros = []
    flotantes = []
    cadenas = []
    booleanos = []

    for elemento in lista:
        if isinstance(elemento, int) and not isinstance(elemento, bool): # Evitar que los booleanos se traten como enteros
            enteros.append(elemento)
        elif isinstance(elemento, float):
            flotantes.append(elemento)
        elif isinstance(elemento, str):
            cadenas.append(elemento)
        elif isinstance(elemento, bool):
            booleanos.append(elemento)

    return enteros, flotantes, cadenas, booleanos

# Lista de datos
datos = [5, 2.4, "hola", 7, 10, 3.14, True, False, "Hi", 0, 0.265, "True", "False"]

# Separar los datos por tipo
enteros, flotantes, cadenas, booleanos = separar_por_tipo(datos)

# Mostrar resultados
print("Enteros:", enteros)
print("Flotantes:", flotantes)
print("Cadenas:", cadenas)
print("Booleanos:", booleanos)


```

#**Ejercicio 8**:

Imprimir pirámide de asteriscos
Descripción: Crea una función que imprima una pirámide de asteriscos usando bucles anidados, donde el número de niveles de la pirámide sea un valor dado.

```python
# Función para imprimir una pirámide
def imprimir_piramide(niveles):
    for i in range(1, niveles + 1):
        print("*" * (2 * i - 1))

# Solicitar el número de niveles al usuario
niveles = int(input("Ingrese el número de niveles para la pirámide: "))

# Imprimir la pirámide
imprimir_piramide(niveles)

```

#**Ejercicio 9**:

Escriba una función que calcule el factorial de un número, el cual debe ser ingresado por el usuario, después usted debe validar que sea un número natural mayor o igual a cero y en caso de que ingrese un número negativo, una letra o cadena, debe pedir nuevamente el número al usuario.

```python
##Ejercicio
def calcular_factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * calcular_factorial(n - 1)

def solicitar_numero():
    while True:
        try:
            numero = int(input("Ingrese un número natural mayor o igual a cero: "))
            if numero < 0:
                print("Por favor, ingrese un número natural mayor o igual a cero.")
            else:
                return numero
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero.")

# Solicitar el número al usuario
numero = solicitar_numero()

# Calcular el factorial
factorial = calcular_factorial(numero)

# Mostrar el resultado
print(f"El factorial de {numero} es {factorial}")
```

#**Ejercicio 10**:
Programar una función que calcule la distancia entre dos puntos en el plano cartesiano,
dadas sus coordenadas (x1, y1) y (x2, y2).

```python
import math

# Función para calcular la distancia entre dos puntos
def calcular_distancia(x1, y1, x2, y2):
    distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distancia

# Solicitar coordenadas al usuario
x1 = float(input("Ingrese la coordenada x1: "))
y1 = float(input("Ingrese la coordenada y1: "))
x2 = float(input("Ingrese la coordenada x2: "))
y2 = float(input("Ingrese la coordenada y2: "))

# Calcular la distancia
distancia = calcular_distancia(x1, y1, x2, y2)

# Mostrar el resultado
print(f"La distancia entre los puntos ({x1}, {y1}) y ({x2}, {y2}) es {distancia:.2f}")
```

#**Ejercicio 11**:

intersectan, son paralelas o coincidentes. Utilice estructuras de control (IF - ELSE)
para manejar los diferentes casos.


```python
# Función para determinar la relación entre dos rectas
def determinar_relacion_rectas(m1, b1, m2, b2):
    if m1 == m2:
        if b1 == b2:
            return "Las rectas son coincidentes."
        else:
            return "Las rectas son paralelas."
    else:
        return "Las rectas se intersectan."

# Solicitar los coeficientes al usuario
m1 = float(input("Ingrese la pendiente (m1) de la primera recta: "))
b1 = float(input("Ingrese la intersección (b1) de la primera recta: "))
m2 = float(input("Ingrese la pendiente (m2) de la segunda recta: "))
b2 = float(input("Ingrese la intersección (b2) de la segunda recta: "))

# Determinar la relación entre las rectas
resultado = determinar_relacion_rectas(m1, b1, m2, b2)

# Mostrar el resultado
print(resultado)
```
